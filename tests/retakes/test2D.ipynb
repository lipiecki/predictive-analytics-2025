## Task 3.
Import dataset from `prices.csv`.

(a) Generate point forecasts with a linear regression model specified by:
$$
\hat{P}_{d} = \beta_0 + \beta_1 P_{d-1} + \beta_2 P_{d-7} + \beta_3 L_{d} + \beta_4 W_{d}
$$
where $P_{d}$ is `price` at day $d$, $L_d$ is `load_da` (load forecast), $W_d$ is `onshore_da` (wind generation forecast). Estimate the model using a 364-day **rolling** window. Print the MAE of the forecasts in the evaluation period from January 1st 2022 to December 31st 2024.

(b) Generate forecasts for the 10% PI **conformal prediction** and **normal distribution** method. Calibrate prediction interval using a 364-day **rolling** window and calculate the following metrics on the evaluation period from January 1st 2024 to December 31st 2024:
- empirical coverage for each method
- p-value from the Kupiec test for each method

based on these results, answer which method produces a better 10% PI.
