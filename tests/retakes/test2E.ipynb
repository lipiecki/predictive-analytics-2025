## Task 3.

$$

$$
where $P_{d}$ is `price` at day $d$, $L_d$ is `load_da` (load forecast), $W_d$ is `onshore_da` (wind generation forecast). Estimate the model using a 364-day **rolling** window. Print the MAE of the forecasts in the evaluation period from January 1st 2022 to December 31st 2024.

(b) Generate forecasts for the 5% and 95% quantile with **historical simulation** and **normal distribution** method. Calibrate prediction interval using a 364-day **rolling** window and calculate the following metrics on the evaluation period from January 1st 2024 to December 31st 2024:
- empirical coverage for each method and quantile
- p-value from the Kupiec test for each method and quantile

based on these results, answer which method produces better quantile forecasts.
