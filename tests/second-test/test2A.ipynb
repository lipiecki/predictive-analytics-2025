## Task 3.
Import dataset from `prices.csv`.

(a) Generate point forecasts with a linear regression model specified by:
$$
\hat{P}_{d} = \beta_1 P_{d-1} + \beta_2 P_{d-7} + \beta_3 L_{d} + \beta_4 W_{d} + \beta_5 S_{d} + \sum_{i=1}^{7} \beta_{i+5} D_{d, i}, $$ 
where $P_{d}$ is `price` at day $d$, $L_d$ is `load_da` (load forecast), $S_d$ is `solar_da` (solar generation forecast), $W_d$ is `onshore_da` (wind generation forecast) and $D_{d,i}$ is the dummy variable for day $d$ corresponding to the $i$-th day of the week. Estimate the model using a 364-day **rolling** window. Print the MSE of the forecasts in the evaluation period from January 1st 2022 to December 31st 2024.

(b) Generate the predictions for 10% and 90% quantile using historical simulation. Calibrate quantile forecasts using a 364-day **rolling** window and calculate the following metrics on the evaluation period from January 1st 2024 to December 31st 2024:
- empirical coverage for each quantile
- p-value from the Kupiec test for each quantile
